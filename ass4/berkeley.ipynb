```python
clocks = [10, 15, 20, 25]
```

```python
print("Clock Times Before Synchronization:")
print(clocks)
```

```python
avg = sum(clocks) / len(clocks)

print("\nAverage Time =", avg)
```

```python
print("\nClock Times After Synchronization:")
```

```python
for i in range(len(clocks)):
    clocks[i] = avg
    print("Processor", i+1, "=", clocks[i])
```
